```python
# Question 1: Rock, Paper, Scissors
import random

# Initial declarations
valid = True
GamesPlayed = 0
PlayerWins = 0
Rock = 0
Paper = 0
Scissors = 0

# Function for computer to choose
def ComputerTurn():
    if (Rock > Paper) & (Rock > Scissors):
        comp = "paper"
    elif (Scissors > Paper) & (Scissors > Rock):
        comp = "rock"
    elif (Paper > Scissors) & (Paper > Rock):
        comp = "scissors"
    elif (Rock == Scissors) & (Rock != Paper):
        comp = random.choice(['rock', 'paper'])
    elif (Scissors == Paper) & (Paper != Rock):
        comp = random.choice(['rock', 'scissors'])
    elif (Paper == Rock) & (Rock != Scissors):
        comp = random.choice(['paper', 'scissors'])
    else:
        comp = random.choice(['rock', 'paper', 'scissors'])
    return (comp)

# Main loop for game
while valid:
    comp = ComputerTurn()
    player = input("Play Rock, Paper, Scissors or Quit?\n")
    player = player.strip().lower()
    if player != "rock" and player != "paper" and player != "scissors" and player != "quit":
        print("Please input a valid option!")
    elif player == "rock":
        Rock += 1
        GamesPlayed += 1
        if comp == "rock":
            print("Both players selected " + comp + " so it's a draw!")
        if comp == "paper":
            print("Computer selected " + comp + " and you selected " + player + " so you lose!")
        if comp == "scissors":
            PlayerWins += 1
            print("Computer selected " + comp + " and you selected " + player + " so you win!")
    elif player == "paper":
        Paper += 1
        GamesPlayed += 1
        if comp == "rock":
            PlayerWins += 1
            print("Computer selected " + comp + " and you selected " + player + " so you win!")
        if comp == "paper":
            print("Both players selected " + comp + " so it's a draw!")
        if comp == "scissors":
            print("Computer selected " + comp + " and you selected " + player + " so you lose!")
    elif player == "scissors":
        Scissors += 1
        GamesPlayed += 1
        if comp == "rock":
            print("Computer selected " + comp + " and you selected " + player + " so you lose!")
        if comp == "paper":
            PlayerWins += 1
            print("Computer selected " + comp + " and you selected " + player + " so you win!")
        if comp == "scissors":
            print("Both players selected " + comp + " so it's a draw!")
    elif player == "quit":
        print("The total number of games played was " + str(GamesPlayed))
        print("The total number of times the human(you) won was " + str(PlayerWins))
        valid = False
```

```python
# Question 3: Near-duplicate detection
import math

# Main function - read in the file and output near duplicates
def main():
    tweets = open("Santa.txt", "r")
    list = []
    whole = []
    for line in tweets:
        whole.append(line)
        list.append(moving_window(line))

    for x in range(len(list)):
        if x < (len(list) - 1):
            for y in range(len(list)):
                if x < y:
                    score = cosine(list[x], list[y])
                    if score > 0.5:
                        print("The tweets " + whole[x] + " and " + whole[y] + " are near-duplicates, with a cosine "
                                                                                 "similarity of " + str(score) + "\n")

                        
# Function to convert a tweet into a dictionary of phrases
def moving_window(tweet):
    tweet = tweet.strip()
    wordlist = tweet.split()
    x = 0
    tweetdict = {}
    for word in wordlist:
        str1 = ""
        if x < len(wordlist) - 2:
            str1 = wordlist[x] + " " + wordlist[x+1] + " " + wordlist[x+2]
            tweetdict[str1] = 1
            x += 1
    return tweetdict

# Function to calculate cosine similarity between two tweets
def cosine(tweet1, tweet2):
    n1 = len(tweet1)
    n2 = len(tweet2)
    matches = 0
    for key in tweet1:
        if key in tweet2:
            matches += 1
    score = matches / (math.sqrt(n1 * n2))
    return score

main()
```

```python
# Question 4: The Google of Quotes

#Part(a) Creating a list of full quotes
import re
fp = open('quotes.txt', 'r')
quoteList = []
for line in fp:
    quoteList.append(line.rstrip())

i = 0
fullQuotes = []
while i < len(quoteList):
    fullQuotes.append(quoteList[i] + ' - ' + quoteList[i+1])
    i+=2
print(fullQuotes)

#Part(b) Function to get words from a full quote
def words_from_quote(fullQuote):
    #print(fullQuote.lower().split(" "))
    fullQuote = (re.sub('`', '', fullQuote))
    wordList = (re.split(r'\W+', fullQuote))
    wordList = [myword.lower() for myword in wordList]
    return wordList
print(words_from_quote(fullQuotes[4])) 

# #Part(c) Creating postings-list
def create_postingsList(fullQuotes):
    postingsList = {}
    for fullQuote in fullQuotes:
        subDict = {}
        wordList = words_from_quote(fullQuote)
        
        for word in wordList:
            if word not in subDict.keys():
                subDict.update({word:wordList.count(word)})
        postingsList.update({fullQuote:subDict})

    return postingsList
print(create_postingsList(fullQuotes))
postingsList = create_postingsList(fullQuotes)

#Part(d) Creating reverse postings-list
def create_revPostingsList(fullQuotes):
    revPostingsList = {}
    uniqueWords = []
    
    for fullQuote in fullQuotes:
        wordList = words_from_quote(fullQuote)
        for word in wordList:
            if word not in uniqueWords:
                uniqueWords.append(word)
    #print(uniqueWords)
    
    for uniqueWord in uniqueWords:
        subDict = {}
        for fullQuote in fullQuotes:
            wordList = words_from_quote(fullQuote)
            if uniqueWord in wordList:
                subDict.update({fullQuote:wordList.count(uniqueWord)})
        revPostingsList.update({uniqueWord:subDict})
    return revPostingsList
        
print(create_revPostingsList(fullQuotes)['entertainer'])
revPostingsList = create_revPostingsList(fullQuotes)

# Part(e) Write a TF-IDF Function
import math

def tf_idf(w, q):
    # TF
    word = postingsList[q][w]
    max_w = max(postingsList[q].values())
    tf = word / max_w
    
    # IDF
    N = len(fullQuotes)
    idf_word = len(revPostingsList[w])
    idf = math.log((N / idf_word))
    
    # TF-IDF
    return(tf * idf)
print(tf_idf("entertainer", "An actor is at most a poet and at least an entertainer. - Marlon Brando"))

# Part(f) Quote search using a single word
def quote_search(word):
    dict = {}
    for quote in revPostingsList[word]:
        dict[quote] = tf_idf(word, quote)
    return dict

print(quote_search("entertainer"))

# Part(g) Quote search using multiple words
def quote_search_multiple(words):
    dict = {}
    uniqueQuotes = []
    for word in words:
        for quote in revPostingsList[word]:
            if quote not in uniqueQuotes:
                uniqueQuotes.append(quote)
                dict[quote] = tf_idf(word, quote)
            else:
                value = dict[quote]
                dict[quote] = tf_idf(word, quote) + value
    print(dict)

quote_search_multiple(['entertainer', 'marlon'])
```
